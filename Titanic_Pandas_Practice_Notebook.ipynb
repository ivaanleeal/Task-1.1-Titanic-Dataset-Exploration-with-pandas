# Explicaci√≥n pd.concat()

Concatena dataframes

```python
#Ejemplo1:
df1 = pd.DataFrame({"Nombre": ["Iv√°n"], "Apellido": ["Leal"]})

concatenado= pd.concat([df,df1], axis=1)
concatenado
```

# Explicaci√≥n de pd.merge()

Combina filas de dos DataFrames comparando los valores de una o m√°s columnas.

```python
#Ejemplo1:

df1 = pd.DataFrame({"PassengerId":[1], "Nombre": ["Iv√°n"], "Apellido": ["Leal"]})

nuevoMerge= pd.merge(df, df1, on="PassengerId", how="inner")


nuevoMerge
```

```python
#Ejemplo2:

df2 = pd.DataFrame({"PassengerId":[1], "Comida": ["Queso"], "Estado": ["Podrido"]})

nuevoMerge2= pd.merge(df, df2, on="PassengerId", how="inner")


nuevoMerge2
```

En el primer ejemplo se crea un dataferame con una columna igual en cada uno de estos y uno de las finas con el mismo valor en esa columna.

En el segundo ejemplo se crea un dataferame con una columna igual en cada uno de estos y uno de las finas con el mismo valor en esa columna.
